#### Example: è¿›åˆ¶è½¬æ¢ (HARD)

````python
def decimal_to_binary(n):  
    binary_num = ''  
    while n > 0:  
        remainder = n % 2
        binary_num = str(remainder) + binary_num # ä¸èƒ½å†™æˆ += é¡ºåºä¸å¯¹
        n //= 2 # n = n // 2
    return binary_num

decimal_number = 10
binary_number = decimal_to_binary(decimal_number)  
print(decimal_number, "çš„äºŒè¿›åˆ¶æ˜¯", binary_number)
````
